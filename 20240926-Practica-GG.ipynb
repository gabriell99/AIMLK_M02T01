### 1. Elección del conjunto de datos

Selecciona un conjunto de datos que despierte tu interés. Pueden ser datos provenientes cualquiera área, como salud, finanzas, deportes, medio ambiente, etc. Recuerda que el conjunto de datos debe ser suficientemente grande para realizar un análisis significativo (mínimo 1000 filas y varias variables) y debe estar disponible en un formato legible por python, como csv, json, excel, etc. Puedes encontrar conjuntos de datos en plataformas como kaggle, uci machine learning repository, o google dataset search.


```python
#Definiendo ruta de acceso
file_path = 'c:/KodigoGG/DataSets/TareasDesarrolloCSV.csv'

#Invocando Libreria Panda
import pandas as pd

#Cargando Archivo
df = pd.read_csv(file_path,sep=';', lineterminator='\r')

#Mostrando Data en Crudo
df.head()
```

### 2. Limpieza y transformación de los datos

Una vez que tengas el conjunto de datos, el siguiente paso es limpiarlo y transformarlo para que esté listo para el análisis. Recuerda verificar las variables, gestionar los valores faltantes, corregir errores y similares

```python
#mostrando el total de registros y columnas
df.shape
```

```python
#Verificando Total de Datos vacios
df.isnull().sum()
```

```python
#Limpiando registros que no tienen Asignado a un responsable (Assigned To)
df = df[(~df['Assigned To'].isnull())]

#Limpiando registros la Hora Total en Negativo
df = df[df['HoraTotal']>=0]

#Limpiando registros si todas columnas están vacias
df.dropna(axis="columns",how="all").shape

#Verificando Total de Datos vacios
df.isnull().sum()

```

```python
#Seleccionando Columnas a Trabajar
df = df[['Nombre', 'State','TDProyectado','FDCumplimiento','HoraTotal','Año', 'Mes' ]]
df.head()
```

```python
#Renombrando Columnas
df.rename(columns={'Nombre':'Responsable'
                   ,'State':'Estado'
                   ,'TDProyectado':'DiasProyectado'
                   ,'FDCumplimiento':'DiasCumplimiento'
                   ,'HoraTotal':'TotalHoras'
                   ,'Año':'Año'
                   ,'Mes':'Mes'}, inplace=True)
```

### 3. Análisis exploratorio de datos (eda)

Ejecuta el análisis exploratorio es esencial para comprender mejor la estructura, las relaciones y las características clave del conjunto de datos generando un resumen estadístico de las variables para identificar su distribución, medias, medianas, desviaciones estándar, etc. Recuerda también identificar patrones y tendencias dentro de los datos.

```python
#Verificando tipos de datos
df.dtypes
```

```python
#Convirtiendo float64 a Int64
df.astype({'DiasProyectado':'int64'}).dtypes
df.astype({'DiasCumplimiento':'int64'}).dtypes
df.astype({'TotalHoras':'int64'}).dtypes
df.astype({'Año':'int64'}).dtypes
df.astype({'Mes':'int64'}).dtypes

df.isnull().sum()
```

```python
#Realizando conteo por "Estado"
df.Estado.value_counts()
```

```python
#Cuentas de las tareas finalizadas se distribuyen en porcentaje por mes
df[df.Estado=="Done"].Mes.value_counts(normalize=True)
```

```python
#Distribucion Estado por Mes
df.groupby(
    "Estado"
).Mes.value_counts(
    normalize=True).unstack()
```

```python
#Distribucion Responsable + Estado por Mes
df.groupby(
    ["Responsable","Estado"]
).Mes.value_counts(
    normalize=True).unstack()
```
