### Tendencias de ventas globales a lo largo de los años

```python
import matplotlib.pyplot as plt

# Agrupar las ventas globales por año
sales_by_year = data.groupby('Year')['Global_Sales'].sum()

# Graficar las ventas globales por año
sales_by_year.plot(kind='line', figsize=(10,6))
plt.title('Tendencia de las ventas globales a lo largo de los años')
plt.xlabel('Año')
plt.ylabel('Ventas globales (en millones)')
plt.show()
```

###  Comparación de ventas por región

```python
# Graficar la comparación de ventas por región
data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum().plot(kind='bar', figsize=(10,6))
plt.title('Comparación de ventas por región')
plt.xlabel('Región')
plt.ylabel('Ventas (en millones)')
plt.show()
```

### Identificación de tendencias por editor o género

```python
# Agrupar las ventas globales por editor
sales_by_publisher = data.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)

# Graficar las ventas por los 10 principales editores
sales_by_publisher.plot(kind='bar', figsize=(10,6))
plt.title('Ventas globales por los principales editores')
plt.xlabel('Editor')
plt.ylabel('Ventas globales (en millones)')
plt.show()
```

### 4. Visualización de resultados
Posteriormente ejecuta la visualización de los datos para comunicar hallazgos y comprender mejor las características del conjunto de datos. Puedes apoyarte de un histograma, un gráfico de barras, de pastel o cualquier otro gráfico que resulte adecuado para los datos seleccionados

Gráfico de Barras: Distribución de géneros de videojuegos

```python
import matplotlib.pyplot as plt

# Gráfico de barras de la distribución de géneros de videojuegos
data['Genre'].value_counts().plot(kind='bar', figsize=(10,6))
plt.title('Distribución de géneros de videojuegos')
plt.xlabel('Género')
plt.ylabel('Cantidad de juegos')
plt.show()
```

###  Gráfico de Pastel: Distribución de ventas globales por los principales editores

```python
# Agrupar las ventas globales por editor
sales_by_publisher = data.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(5)

# Gráfico de pastel para los 5 principales editores por ventas
sales_by_publisher.plot(kind='pie', autopct='%1.1f%%', figsize=(8,8))
plt.title('Distribución de ventas globales por los 5 principales editores')
plt.ylabel('')  # Ocultar la etiqueta del eje Y
plt.show()
```

### Boxplot: Ventas globales por plataforma

```python
# Boxplot de ventas globales por plataforma
data.boxplot(column='Global_Sales', by='Platform', figsize=(12,8))
plt.title('Distribución de ventas globales por plataforma')
plt.suptitle('')  # Eliminar el título automático
plt.xlabel('Plataforma')
plt.ylabel('Ventas globales (en millones)')
plt.xticks(rotation=90)
plt.show()
```

### Scatter Plot: Relación entre ventas en Norteamérica y Europa

```python
# Gráfico de dispersión entre ventas en Norteamérica y Europa
data.plot(kind='scatter', x='NA_Sales', y='EU_Sales', alpha=0.5, figsize=(10,6))
plt.title('Relación entre ventas en Norteamérica y Europa')
plt.xlabel('Ventas en NA (en millones)')
plt.ylabel('Ventas en EU (en millones)')
plt.grid(True)
plt.show()
```
